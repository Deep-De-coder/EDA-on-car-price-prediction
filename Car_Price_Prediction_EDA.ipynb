```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (16, 8)
plt.style.use('fivethirtyeight')

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data.csv')
display(df.head())
print('Shape of the Data: ', (df.shape))
```

```python
df.describe()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head()
```

```python
categorical = list(df.dtypes[df.dtypes == 'object'].index)
```

```python
for col in categorical:
    df[col] = df[col].str.lower().str.replace(" ", "_")
    
df.head()
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique()[:5])
    print(df[col].nunique())
    print('\n')
```

```python
print("Number of Missing Values in our data set\n")
missing_df = df.isnull().sum().to_frame().reset_index().rename({"index" : 'Variable', 0: 'Missing Values'}, axis =1)
display(missing_df.style.background_gradient('gnuplot2_r'))
print("\n Percentage of Missing Values in our data set")
display((df.isnull().sum() / (len(df.index)) * 100).head(20).to_frame().rename({0:'Count'}, axis = 1).style.background_gradient('gnuplot2_r'))
round((df.isnull().sum() / (len(df.index)) * 100) , 2).plot(kind = 'barh',color ='#bf0606')

plt.title("Percentage of Missing values");
```

```python
df['market_category'].fillna(df['market_category'].mode()[0], inplace = True)
df['engine_fuel_type'].fillna(df['engine_fuel_type'].mode()[0], inplace = True)

df['engine_hp'].fillna(df['engine_hp'].mean(), inplace = True)

df['engine_cylinders'].fillna(df['engine_cylinders'].median(), inplace = True)
df['number_of_doors'].fillna(df['number_of_doors'].median(), inplace = True)

display(df.isnull().sum().to_frame().reset_index().rename({'index' : 'Variables', 0: 'Missing Values'}, axis =1).style.background_gradient('copper_r'))
```

```python
display("Total number of of Duplicates present in data: %s" %df.duplicated().sum())
```

```python
df.drop_duplicates(inplace = True)

print("Total number of of Duplicates present in data: %s" %df.duplicated().sum())
```

```python
df.corr().style.background_gradient('copper_r')
```

```python
sns.heatmap(df.corr(), cmap = 'copper_r', annot = True, lw = 0.1);

```

```python
numerical = [col for col in df.columns if col not in categorical]
for i in numerical:
    ax = sns.distplot(df[i], color = '#bf0606')
    plt.title("Distribution of %s" %i, fontsize = 20)
    plt.xlabel(" ")
    plt.ylabel(" ")
    plt.xticks(fontsize = 15)
    plt.show();
    print('\n')
```

```python
nrows = 2
ncols = 4
i = 0
fig, ax = plt.subplots(nrows, ncols, figsize = (16,8),)
for row in range(nrows):
    for col in range(ncols):
        sns.scatterplot(df[numerical].iloc[:, i],df['msrp'],ax = ax[row, col],color = '#bf0606' ).set(ylabel = '')
        plt.tight_layout()
        i += 1   
```

```python
def sort_order(column):
    orders = (df.groupby([column]).mean().sort_values(by ='msrp', ascending = False)).index
    return orders
```

```python
for i in categorical:
    if df[i].nunique() < 72:
        sns.barplot(df[i],df['msrp'], order = sort_order(i), palette='copper')
        plt.title("Bar Plot of %s" %i, fontsize = 20)
        plt.xticks(fontsize = 12)
        plt.xlabel("%s"%i)
        plt.ylabel("Car Price")
        plt.xticks(fontsize = 15, rotation = 90)
        plt.show();
        print('\n')
```
